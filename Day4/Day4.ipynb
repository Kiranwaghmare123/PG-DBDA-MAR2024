# Regression

```python
# 1. Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#2 Import dataset
dataset = pd.read_csv('Day4.csv')
```

```python
dataset
```

```python
#3. Define variables
#Independent variable
x = dataset.iloc[:,:-1].values

#Dependent variable
y = dataset.iloc[:, -1].values
```

```python
#x
#y
```

```python
#4. Splitting dataset for Training and Testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=.2,random_state=1)
```

```python
#5. Model Building
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
#6. Prediction
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
#7.Visualizing the training dataset
plt.scatter(X_train,y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Simple Linear Regression")
plt.xlabel("Salary")
plt.ylabel("Experience")
```

```python
#8.Visualizing the testing dataset
plt.scatter(X_test,y_test, color='red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title("Simple Linear Regression")
plt.xlabel("Experience")
plt.ylabel("Salary")
```

```python
#Assigning coefficient
b = regressor.coef_
b
```

```python
#Assigning y-intercept value
a = regressor.intercept_
a
```

```python
# Prediction
print(regressor.predict([[15]]))
```

```python
from sklearn import metrics
metrics.mean_squared_error(y_test,y_pred)
```

# Boston House Price Prediction 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
boston
```

```python
boston
```

```python
X =pd.DataFrame(boston.data, columns=boston.feature_names)
X
```

```python
y = boston.target
y
```

```python
#adding constant for intercept term
X = sm.add_constant(X)
```

```python
#Split the dataset into training and testing
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
import statsmodels.api as sm
model = sm.OLS(y_train,X_train).fit()
```

```python
model.summary()
```

```python
#Calculate statistics
dw_statistics = sm.stats.durbin_watson(model.resid)
dw_statistics 
```
