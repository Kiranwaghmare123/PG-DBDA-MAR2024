```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2, markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
```

```python
#Subplot creation
plt.subplot(2,2,1)
plt.plot(x,y,'r--')

plt.subplot(2,2,2)
plt.plot(x,y,'g*-')

plt.subplot(2,2,3)
plt.plot(x,y,'bo')

plt.subplot(2,2,4)
plt.plot(x,y,'go')
```

```python
x = np.arange(1,11)
y = 3 * x + 5
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Matplotlib Graph in 2D')
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
#Bar plot
x=[2,8,10]
y=[11,16,9]
plt.bar(x,y, color ='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Barplot Graph in 2D')

```

```python
#Bar plot
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y, color ='g')
plt.bar(x2,y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Barplot Graph in 2D')

```

```python
#Histogram
a=np.array([2,67,89,34,67,23,12,78.89,34,90,45,78])
plt.hist(a)
plt.title('Histogram')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
#Boxplot
data = [np.random.normal(0,std,100) for std in range(1,4)]
plt.boxplot(data,vert=True,patch_artist=False)
```

```python
#Pie char
labels = 'Python','C++','Java','Ruby'
sizes = [215,130,245,210]
colors = ['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0 ,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%1.1f%%',shadow=False)
plt.axis('equal')
plt.show()
```
